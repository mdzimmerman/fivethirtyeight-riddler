```python
import networkx as nx
import matplotlib.pyplot as plt

import re
```

```python
G = nx.Graph()
```

```python
with open('states.txt', 'r') as fh:
    for l in fh:
        l = re.sub('#.*\n', '', l)
        l = l.rstrip()
        if l == '':
            continue
        s = l.split(',')
        #print(s)
        src = s[0]
        #G.add_node(src)
        for dest in s[1:]:
            G.add_edge(src, dest)
            #print(src, dest)
```
