```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d = np.arange(0.01, 1, 0.01)
```

```python
r = 1/d
```

```python
df = pd.DataFrame({
    "D": d,
    "r": r,
    "r2": np.round(r, 2)
})
df
```

```python
df['delta'] = 100 * df.D * df.r2 - 100
df
```

```python
plt.plot(df.D, df.delta)
plt.show()
```

```python
df.sort_values(by=["delta"], ascending=False)[:10]
```
