```python
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import fsolve
```

```python
def subfig(ax, r):
    c1 = plt.Circle((0, 0), 1, fill=0)
    c2 = plt.Circle((1, 0), r, fill=0)
    ax.set_xlim(-1.5, 2.5)
    ax.set_ylim(-2, 2)
    ax.add_patch(c1)
    ax.add_patch(c2)
    ax.set_aspect('equal')

fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3)
axs = [ax1, ax2, ax3, ax4, ax5, ax6]
rs  = [0.1, 0.5, 1.0, 1.15, 1.5, 1.9]
for ax, r in zip(axs, rs):
    subfig(ax, r)
```

```python
def A(r):
    return (r**2)*math.acos(0.5*r) + math.acos((2.0-r**2)/2.0) - 0.5*math.sqrt((r**2)*(2-r)*(2+r))
```

```python
A(0.5)
```

```python
rr = np.arange(0, 2.01, 0.01)
plt.plot(rr, [A(r) for r in rr], '-')
#plt.plot(rr, [A(r) - math.pi/2 for r in rr], '-')
plt.show()
```

```python
f = lambda r : A(r) - math.pi/2
```

```python
r0 = fsolve(f, 1.0)
```
