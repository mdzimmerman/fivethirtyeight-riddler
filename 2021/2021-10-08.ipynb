```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import minimize
```

```python
prize = 10_000_000
n_other = 10

def profit(n):
    return ((prize/(n + n_other))*n - n)
    
def profit_min(n):
    return prize - profit(n)
```

```python
n = np.arange(1, 1000000)
plt.plot(n, profit(n))
plt.plot(n, profit_min(n))
plt.xscale('log')
#plt.yscale('log')
plt.show()
```

```python
res = minimize(profit_min, 1000, method='nelder-mead',
               options={'xatol': 1e-8, 'disp': True})
res.fun
```
