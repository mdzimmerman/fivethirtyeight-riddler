```python
class Block:
    def __init__(self, size, collapse_order=None):
        self.size=size
        self.houses=np.full(size, 1)
        if collapse_order!=None:
            self.collapse_order=collapse_order
        else:
            self.collapse_order=list(range(size))
            random.shuffle(self.collapse_order)
        self.round=0
        self.states=[]
        self.states.append(list(self.houses))
        
    def isa_neighbor_collapsed(self, i):
        if (i==0):
            return self.houses[i+1]==0
        elif (i==len(self.houses)-1):
            return self.houses[i-1]==0
        else:
            return self.houses[i-1]==0 or self.houses[i+1]==0

    def decay(self):
        while (self.houses == 1).sum() > 0:
            new_collapses=[]
            for i in range(self.size):
                if self.houses[i]==1 and self.isa_neighbor_collapsed(i):
                    new_collapses.append(i)
            for j in new_collapses:
                self.houses[j]=0
            self.houses[self.collapse_order[self.round]]=0
            self.states.append(list(self.houses))
            self.round+=1
            
    def print_state(self):
        for s in self.states:
            print("".join([str(x) for x in s]))
    
results=[]
for i in range(1000):
    b=Block(36)
    b.decay()
    #if (b.round>=20):
        #b.print_state()
        #print()
    results.append(b.round)
plt.hist(results,bins=list(range(30)))
plt.show()
```
